```
df = spark.table("df6_csv")
display(df)
# Column 'label' indicates whether the return of sp500 is positive or negative
```

```
col_list = df.columns
col_list.remove('Date')
col_list
```

```
# Assign the schema to the table
df = df.withColumn('Date', df['Date'].cast("timestamp"))
for col in col_list:
  df = df.withColumn(col, df[col].cast("double"))
display(df)
```

```
# Construct a feature vector by combine the feature values
from pyspark.ml.feature import VectorAssembler
assembler = VectorAssembler(inputCols=['dji', 'hsi', 'n225', 'aord'], outputCol="features")
df = assembler.transform(df)
display(df)
```

```
# Split to training and testing dataset
from pyspark.sql.functions import percent_rank
from pyspark.sql import Window
df = df.withColumn("rank", percent_rank().over(Window.partitionBy().orderBy("Date")))
train = df.where("rank <= .75").drop("rank")
test = df.where("rank > .75").drop("rank")
print("Training Dataset Count: " + str(train.count()))
print("Test Dataset Count: " + str(test.count()))
```

```
from pyspark.ml.classification import RandomForestClassifier
```

```
# Model training and prediction
rf = RandomForestClassifier(featuresCol = 'features', labelCol = 'label')
rfModel = rf.fit(train)
predictions = rfModel.transform(test)
predictions.select('dji', 'hsi', 'n225', 'aord', 'label', 'prediction').show(10)
```

```
# Evaluate using AUC
from pyspark.ml.evaluation import BinaryClassificationEvaluator

evaluator = BinaryClassificationEvaluator()
print("Test Area Under ROC: " + str(evaluator.evaluate(predictions, {evaluator.metricName: "areaUnderROC"})))
```

```
# Evaluate using Accuracy
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
evaluator = MulticlassClassificationEvaluator(metricName="accuracy")
print("The accuracy: " + str(evaluator.evaluate(predictions)))
```

```
display(predictions)
```

```
# Construct the trading strategy based on the model

from pyspark.sql import functions as F

my_window = Window.partitionBy().orderBy("Date")
predictions = predictions.withColumn("sp500_Price_Lag", F.lag(predictions['sp500_Price']).over(my_window))
predictions = predictions.withColumn("Price_Change", F.when(F.isnull(predictions['sp500_Price'] - predictions['sp500_Price_Lag']), 0).otherwise(predictions['sp500_Price'] - predictions['sp500_Price_Lag']))
display(predictions)
```

```
predictions = predictions.dropna()
display(predictions)
```

```
predictions = predictions.withColumn("profit", F.when(predictions['prediction']==1, predictions['Price_Change']).otherwise(predictions['Price_Change'] * -1))
predictions = predictions.withColumn("wealth", F.sum('profit').over(my_window) + predictions.collect()[0]['sp500_Price'])
display(predictions)
```
